< 누적합(Prefix Sum) >

배열(리스트)의 특정 구간의 합을  빠르게 구하기 위해 사용하는 알고리즘 

```python
nums =[1, 3, 5, 7, 9, 10, 20, 10]

# 전체 총합
t1 = 0

for i in range(len(nums)):
    t1 += nums[i]

print(t1)

# 1번째부터 3번째까지의 합
t2 = 0

for i in range(3): # 0~2
    t2 += nums[i]
# 이전에 물었던 구간과 반복구간이 있어도 다시 연산을 반복하여야 한다.
print(t2) 
```

```python
# 누적합 배열 만들기

nums =[1, 3, 5, 7, 9, 10, 20, 10]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1] + n) # 누적합 값인 원소인 마지막 원소에 다음 원소값을 누적해서 append
    # 0+1->10 append [0, 1]
    # 1+3->4 append [0, 1, 4]
    # 4+5->9 append [0, 1, 4, 9]
    # ...  [0, 1, 4, 9, 16, 25, 35, 55, 65]
    
print(prefix)
print(prefix[3]) 
print(prefix[7])
```

```python
# 구간 합 공식
# 구간합 = P[end] - P[start-1]

nums = [10, 2, 30, 21 , 44, 5, 8, 50, 7]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1] + n)

# 주의 : 구간의 숫자는 인덱스가 아니다
# 3번째(값 30)에서 7번째(값 8)까지의 합
start = 3
end = 7

result = prefix[end] - prefix[start - 1]

print(result)
```

```python
# 실습
scores = [80, 90, 75, 88, 95, 70, 85]

prefix = [0]

for i in scores:
    prefix.append(prefix[-1] + i)
    
print(prefix) # [0, 80, 170, 245, 333, 428, 498, 583]

# 1~3번 학생 총점?
print(prefix[3] - prefix[1-1])

# 2~5번 학생 총점은? 
print(prefix[5] - prefix[2-1])

# 3~7번 학생 총점은? 
print(prefix[7] - prefix[3-1])

# 1~7번 학생 총점은?
print(prefix[7] - prefix[1-1])
```
