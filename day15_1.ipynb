```python
texts = ["abc님 안녕하세요", "axc님 안녕하세요"]
for t in texts:
    r = t.replace("abc", "quest")
    print(r)

# axc는 변하지 않음

for t in texts:
    if t.startswith("abc"):
        r = t.replace("abc", "guest")
    if t.startswith("axc"):
        r = t.replace("axc", "guest")   
    print(r)

# 데이터가 많아질수록 코드가 길어지고 복잡해짐
```

```python
import re
texts = ["abc님 안녕하세요", "axc님 안녕하세요"]
for t in texts:
    r = re.sub(r"a.c", "quest", t)
    print(r)

# re 모듈을 사용할경우 간단하게 표현가능
# 구체적인 값 문자열을 찾을때는 문자열 메서드를 사용하는 것이 편할 수도 있지만, 특정 패턴을 가진 문자열들을 찾을때는 re모듈의 함수를 사용하는 것이 좋다.
```

```python
# re.match(패턴, 문자열)

import re
text = "Python"
result = re.match(r"Py", text)
print(result)
import re

text = "Python"
result = re.match(r"th", text)
print(result) # None, match()는 문자열의 시작에서만 검사
```

```python
# re.search() 

import re

text = "Python"
print(re.search(r"th", text)) # <re.Match object ...>
```

```python
# re.findall() 모든 검색 결과를 리스트로 반환한다.

import re

text = "apple banana orange"
result = re.findall(r"a", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
import re

text = "A10 B20 C30"
numbers = re.findall(r"\d+", text) # \d 숫자한문자, \d+ 연속된 숫자 한문자이상
print(numbers)
```

```python
# re.sub(패턴, 변경할문자, 원본문자열)

# 숫자 문자 제거
import re

text = "A10 B20 C30"
result = re.sub(r"\d", "", text)
print(result)
result2 = re.sub(r"[^\d\s]", "0", text) # 숫자와 공백은 제외
print(result2)
```

* match() : 문자열의 시작부분과 패턴이 일치하는지 검사
* search() : 문자열 전체에서 패턴이 존재하는지 검사, 처음 발견되는 곳 반환
* findall() : 문자열 전체에서 패턴과 일치하는 모든 결과를 리스트로 반환
* finditer() :  문자열 전체에서 패턴과 일치하는 모든 결과를 Match 객체의 반복자로 반환
* sub() : 문자열 치환, 원본은 변하지 않음
* split() : 패턴 기준으로 문자열 분리
