< 슬라이딩 윈도우(Sliding Window) >

슬라이딩 윈도우(Sliding Window)는 배열이나 리스트에서 고정된 크기의 창(Window)을 옆으로 밀면서(Sliding) 문제를 해결하는 효율적인 알고리즘이다.
start와 end 2개의 두 점 사이의 간격이 일정하게 유지되면서 이동하는 모습이 마치 창문을 이동시키는 모양새와 같다고 해서 슬라이딩 윈도우라는 이름으로 불리게 되었다.

```python
scores = [10, 20, 30, 40, 50, 60]

k = 3

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[:k]) # scores[0:3]

print(window)

# 초기값에서 1칸을 이동 → k까지

for i in range(k, len(scores)):  # range(3,6)
    window = window - scores[i-k] + scores[i] 
    # [10+20+30] - [10] + [40]
    # [20+30+40] - [20] + [50]
    # [30+40+50] - [30] + [60]
    print(window) # 90

# 슬라이딩 윈도우 : "고정 크기의 구간을 한 칸씩 이동하면서 이전 계산 결과를 재사용하는 알고리즘"
```

```python
scores = [10, 20, 3, 40, 5, 60, 9, 55, 23]

k = 5

window = sum(scores[:k]) # scores[0:3]

print(window)

for i in range(k, len(scores)):  # range(3,6)
    window = window - scores[i-k] + scores[i] 
    print(window) 
```

```python
# 윈도우 구간 평균 구하기

scores = [80, 90, 70, 100, 95, 85]

k = 3

window = sum(scores[:k])

print(window / k)

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]
    print(round(window / k, 2))
```

```python
# 윈도우 구간 합의 최대값(최대합) 구하기

scores = [10, 20, 30, 40, 50, 60]

k = 3

window = sum(scores[:k])

max_sum = window

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]

    if window > max_sum:
        max_sum = window

print(max_sum)
```

```python
scores = [33, 63, 48, 105, 73, 123, 107]
k = 3

window = sum(scores[:k])

total = [window]
avg = [window/k]
max_val = window 
min_val = window 

print(window)

for i in range(k,len(scores)):
    window = window - scores[i-k] + scores[i]
    total.append(window)
    avg.append(round(window/k, 2))

    if window > max_val:
        max_val = window

    if window < min_val:
        min_val = window

print(total)
print(avg)
print(max_val)
print(min_val)
```

```python
scores = [33, 63, 48, 105, 73, 123, 107]
k = 3

window = sum(scores[:k])
print(window) # 144

total = [window]
avg = [window/k]
max_val = {"구간" : f"0~{k-1}", "합" : window}
min_val = {"구간" : f"0~{k-1}", "합" : window}


for i in range(k,len(scores)):
    window = window - scores[i-k] + scores[i]
    total.append(window)
    avg.append(round(window/k, 2))

    section =f"{i-(k-1)}~{i}" # f"{i-k+1}~{i}"

    if window > max_val["합"]:
        max_val = {"구간" : section, "합" : window}

    if window < min_val["합"]:
        min_val = {"구간" : section, "합" : window}

print(total)
print(avg)
print(max_val)
print(min_val)
```

슬라이딩 윈도우는 "연속적인 데이터 스트림"이나 "고정된 범위의 상태를 유지해야 할 때" 주로 사용 된다.

1. 네트워크 패킷 전송 (TCP Sliding Window): 네트워크에서 데이터를 보낼 때, 수신 측이 처리할 수 있는 만큼의 '윈도우 크기'를 정해두고, 데이터가 성공적으로 도착하면 윈도우를 옆으로 밀어가며 다음 데이터를 연속해서 전송한다. (흐름 제어의 핵심 기술)

2. 시계열 데이터의 이동평균(Moving Average) 계산: 주식 시장에서 '5일 이동평균선', '20일 이동평균선'을 구할 때 매일매일 지난 5일치 데이터를 새로 더하는 게 아니라, 슬라이딩 윈도우 방식으로 가장 오래된 날짜는 빼고 오늘 날짜는 더하며 빠르게 계산한다.​

3. 문자열 매칭 및 패턴 탐색: 긴 문장 속에서 "길이 4짜리 특정 DNA 염기서열(예: AGCT)이 포함되어 있는가?"를 찾기 위해 글자 4개짜리 창문을 만들어 오른쪽으로 한 글자씩 밀어가며 탐색한다.

4. 실시간 데이터 스트리밍 분석: 최근 10분간의 실시간 트래픽 합계, 최근 1시간 동안의 웹사이트 방문자 수 변화 등을 계속 업데이트하며 모니터링 분석할 때 사용된다.
