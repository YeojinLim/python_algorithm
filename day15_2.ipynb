```python
# 실습1) 상품명에서 가격만 출력

import re

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    price = re.search(r"\d+", product)
    print(price.group())
    price_list.append(price.group())

print(price_list)
```

```python
import re

products = [
    "사과 3500원 555",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    name = re.search(r"[가-힣]+",product)
    price = re.search(r"\d+", product)
    price_list.append({name.group() : int(price.group())})

print(price_list)
```

```python
# 실습2) 여러 공백을 하나의 공백으로 치환

import re

text = "Python   Data   Analysis"
result = re.sub(r"\s+", " ", text) # \s+ 1개이상의 공백 문자
print(result)
```

```python
# 실습3) 이메일 검증

import re

emails = [
    "python@test.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
# 실습4) URL 검증

import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"

# ^https? →→→→→ ^http : http라는 글자로 시작해야한다. / s? : s가 있어도 되고 없어도 됨 < "http" or "https" 로 시작 >
# :// →→→→→ :// 가 그대로 와야한다. < "http://" or "https://" 로 시작 >
# [\w-] →→→→→ []: 이 안에 있는 글자 중 아무 한 글자/ \w- : 알파벳, 숫자, 밑줄 / - : 하이픈 < 알파벳, 숫자, 밑줄(_), 혹은 하이픈(-)" 중 아무글자 한개 >
# +: 글자 연속 반복 가능 / "\." : 마침표 점
# [\w-]+ →→→→→ 영문, 숫자, 하이픈 등이 1글자 이상 온다 < https://abc, https://ab-c, https://-a-, https://abc.abc.abc 등 >
# (/[^\s]*)? →→→→→ / : 순수 슬래시 기호, ^ : 제외, \s : 공백 , * : 글자가 0개 이상 여러 개 올 수 있음 , ? : 있어도 되고 없어도됨 < 상세 페이지 경로 >
# $ : 끝

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
# 실습5) 주민등록변호 형식 검사(외국사람 제외)

import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

# {n} : 앞글자 혹은 패턴 n번 반복
# ^\d{6} : 숫자 6개로 시작
# - : 순수 하이픈 기호 있어야함
# [1-4] : 1부터 4까지의 숫자중 아무거나 한개
# \d{6}$ :  숫자 6개로 끝


for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
