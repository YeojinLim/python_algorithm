```python
class Product:
    
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count

    def print_info(self):
        print(f"상품명 : {self.name}")
        print(f"가격 : {self.price:,}원")
        print(f"재고: {self.stock}개")

p = Product("키보드", 50000, 10)

# print(p.name)
# print(p.price)
# print(p.stock)
p.print_info()
```

실습 6. 메서드를 이용한 기능 구현

- 다음 메서드를 추가한다.
  order(count)
  total_price(count)
  is_sold_out()

- 기능

  주문하면 재고 감소 
  주문 금액 계산 
  품절 여부 반환 


출력
- 키보드 2개 주문
- 결제금액 : 100000원
- 남은 재고 : 8개

추가 문제
- 주문 수량이 재고보다 많으면 재고 부족 을 출력한다.

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def order(self, count):
        if count > self.stock:
            print("재고 부족")
        else:
            self.stock -= count
            print(f"- {self.name} {count}개 주문")
            print(f"- 결제금액 : {self.total_price(count)}원")
            print(f"- 남은 재고 : {self.stock}개")
    
    def total_price(self, count):
        return self.price * count
    
    def is_sold_out(self):
        if self.stock == 0:
            print("품절입니다.")
        else:
            print(f"품절이 아닙니다. 재고 : {self.stock}개")

products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3)
]

products[0].order(2)

# 주문 수량이 재고보다 많은 경우
print("-"*20)
products[0].order(10)
```

의 메서드 형태로 변경 한다.​

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def is_sold_out(self):
        return self.stock == 0

    def print_info(self):
        print(f"상품명 : {self.name} | 가격 : {self.price}원 | 재고 : {self.stock}개")


# ==================================================
# 💡 [실습 7 - Step 3] 상품들을 통틀어 관리하는 매니저 클래스
# ==================================================
class ProductManager:
    def __init__(self):
        self.products = []

    # 1. 상품 등록 기능
    def add_product(self, product):
        self.products.append(product)

    # 2. 전체 상품 조회 기능
    def show_all(self):
        print("\n--- [전체 상품 목록] ---")
        for p in self.products:
            p.print_info()

    # 3. 상품 검색 기능 (이름으로 찾기)
    def search_product(self, name):
        for p in self.products:
            if p.name == name:
                return p  # 찾으면 그 상품 객체 자체를 반환
        return None  # 못 찾으면 아무것도 없다고 반환

    # 4. 품절 상품 조회 기능
    def show_sold_out(self):
        print("\n--- [품절 상품 목록] ---")
        for p in self.products:
            if p.is_sold_out():  # 6번에서 만든 기능 확인!
                print(f"{p.name} (품절)")

    # 5. 가장 비싼 상품 찾기 기능
    def get_most_expensive(self):
        if not self.products:
            return None
        return max(self.products, key=lambda x: x.price)

    # 6. 재고가 가장 많은 상품 찾기 기능
    def get_max_stock(self):
        if not self.products:
            return None
        return max(self.products, key=lambda x: x.stock)

    # 7. 전체 상품 재고 수량 합계 기능
    def get_total_stock_count(self):
        return sum(p.stock for p in self.products)

    # 8. 전체 재고 금액(가격 × 재고) 계산 기능
    def get_total_stock_value(self):
        return sum(p.price * p.stock for p in self.products)

manager = ProductManager()

manager.add_product(Product("키보드", 50000, 10))
manager.add_product(Product("마우스", 30000, 20))
manager.add_product(Product("모니터", 250000, 5))
manager.add_product(Product("노트북", 1200000, 0))

# 전체 상품 조회
manager.show_all()

# 상품 검색 후 조작
print("\n--- [상품 검색 및 수정] ---")
found_product = manager.search_product("마우스")
if found_product:
    print(f" 검색 성공-> {found_product.name}")
    found_product.price = 25000  # 가격 변경
    print(f"{found_product.name}의 가격을 {found_product.price}원으로 내렸습니다.")

# 품절 상품 조회
manager.show_sold_out()

# 가장 비싼 상품 & 재고 최고 상품 찾기
expensive = manager.get_most_expensive()
rich_stock = manager.get_max_stock()

print("\n--- [매장 통계 분석] ---")
print(f"가장 비싼 상품: {expensive.name} ({expensive.price}원)")
print(f"재고가 가장 많은 상품: {rich_stock.name} ({rich_stock.stock}개)")

# 총 수량 및 총 금액 계산
print(f"총 재고 수량: {manager.get_total_stock_count()}개")
print(f"총 재고 금액: {manager.get_total_stock_value()}원")
```
