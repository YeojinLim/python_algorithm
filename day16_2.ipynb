1. 객체지향 패러다임

- 딕셔너리는 데이터를 저장하는 자료구조이고, 클래스는 데이터를 저장하는 속성과 그 데이터를 처리하는 메서드를 하나로 묶어 객체를 만드는 설계도이다. 
이러한 객체를 중심으로 프로그램을 설계하는 방식이 객체지향 프로그래밍(Object-Oriented Programming, OOP)이며, 이러한 사고방식을 객체지향 패러다임이라고 한다. 

​< 객체지향 프로그래밍(Object-Oriented Programming, OOP) 특징 >

- 객체(Object) 중심으로 프로그램을 설계한다.
데이터가 아닌 '학생', '자동차', '은행계좌' 같은 객체를 중심으로 생각한다. 

- 데이터와 기능을 하나로 묶는다. (캡슐화)
속성(데이터)과 메서드(기능)를 하나의 객체로 관리한다. 

- 코드 재사용이 쉽다.
클래스를 여러 번 생성(객체 생성)하여 사용할 수 있다. 

- 상속(Inheritance)이 가능하다.
기존 클래스를 물려받아 새로운 클래스를 만들 수 있다. 

- 유지보수가 쉽다.
기능이 변경되면 해당 클래스만 수정하면 되므로 수정 범위가 줄어든다. 

- 확장이 쉽다.
새로운 기능이나 클래스를 기존 코드에 큰 영향을 주지 않고 추가할 수 있다. 

- 현실 세계를 모델링하기 쉽다.
사람, 자동차, 학생, 동물처럼 실제 사물을 객체로 표현하기에 적합하다.

```python
class Student:
    def __init__(self, name):
        self.name = name
        
    def print_name(self):
        print(self.name)
    
    def greet(self):
        print(f"{self.name}님 안녕하세요")

# 상속
class HighStudent(Student):
    def study(self): # study 라는 매서드 추가
        print("공부한다.") 

# 사용
s = HighStudent("홍길동")
s.print_name()
s.study()
s.greet()
```

```python
# 1) 딕셔너리 함수

def average(student):
    return (student["kor"] + student["eng"]) / 2

student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "average": average
}

print(student["average"](student))

student2 = {
    "name": "김철수",
    "kor": 80,
    "eng": 70,
    "average": average
}

print(student2["average"](student2))
 
# 데이터 객체(name, kore, emg), 함수 객체(average)가 각각 따로 존재하고, 딕셔너리는 단순히 둘을 저장하고 있을 뿐이다.
# 함수에는 학생에 대한 정보가 없기 때문에 호출할때 학생 인수로 학생의 정보를 전달해야만 한다.
```

```python
# 2) 클래스 메서드

class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def average(self):
        return (self.kor + self.eng) / 2


s = Student("홍길동", 90, 85)
s2 = Student("김철수", 80, 70)

print(s.average())
print(s2.average())

# 메서드는 자기가 다루어야 할 학생의 정보(자신의 객체 : self)를 이미 알고 있다.
# 메서드는 객체 내부에 소속되어 있으며, 자신의 데이터(self)를 직접 사용할 수 있다.
# 딕셔너리도 함수를 값으로 저장할 수 있지만, 함수는 딕셔너리에 소속된 것이 아니라 독립적인 객체이다. 따라서 데이터를 항상 인수로 전달해야 한다. 
# 반면 클래스의 메서드는 객체에 소속되어 self를 통해 자신의 데이터를 직접 사용할 수 있다는 점이 객체지향의 핵심 차이이다.
```

```python
# 딕셔너리만 사용

student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85
}

print(student["name"])
print(student["kor"] + student["eng"])

s1 =  {"name":"홍길동", "kor":90, "eng":85}
s2 =  {"name":"이순신", "kor":100, "eng":95}
s3 =  {"name":"김유신", "kor":80, "eng":90}

a1 = (s1["kor"] + s1["eng"]) / 2
a2 = (s2["kor"] + s2["eng"]) / 2
a3 = (s3["kor"] + s3["eng"]) / 2

print(f"{s1["name"]} 평균 점수 : {a1}")
print(f"{s2["name"]} 평균 점수 : {a2}")
print(f"{s3["name"]} 평균 점수 : {a3}")

# 학생이 여러명이면 코드가 계속 반복이 많고, 만약 평균 계산 방법이 바뀌면 모든 곳을 수정해야 한다.
```

```python
# 클래스로 변경시

class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = 0


    def total(self):
        return self.kor + self.eng + self.math

    def average(self):
        return self.total() / 3


s1 = Student("홍길동", 90, 85)
s1.math = 99
print(f"{s1.name} 평균 점수 : {s1.average():.2f}")

s2 = Student("김철수", 80, 88)
s2.math = 65
print(f"{s2.name} 평균 점수 : {s2.average():.2f}")

# math 라는 변수가 추가되면 클래스 내부의 메서드 코드만 변경하면 된다.
```
