1. 클래스

클래스는 < 사용자 정의 데이터 타입 >이다.
클래스는 객체를 만들기 위한 설계도(틀) 이다(객체지향 프로그래밍(OOP)의 핵심 개념)
* 기존의 딕셔너리 타입에 객체지향 패러다임을 추가한 데이터 타입 

2. 클래스 기본 구조

* ​class : 클래스 정의 키워드
* __init__ : 생성자 (객체 생성 시 자동 실행)
* self : 자기 자신 객체를 의미(자바의 this와 동일)
* def : 메서드 정의

class 클래스이름:

    def __init__(self, 매개변수):
        self.속성 = 값

    def 메서드이름(self):
        실행코드

* 메서드(멤버함수) vs 함수

- 메서드는 특정 클래스로 만든 하나의 객체(데이터)에만 종속되는 기능으로 클래스 소스 코드안에 같이 정의 되어 있어야한다.
- 함수는 특정 클래스로 만든 모든 객체(데이터)에 사용가능한 기능으로 클래스와 별개로 기능을 사용하는 곳에 함수가 있어야 한다.

ex) 
메서드 : 하나의 TV에만 사용가능한 리모콘으로 TV를 구매하면 포함되어 있으며 이 TV에만 (종속)사용 가능하다.
함수 : TV와는 별개로 구매해야 하고, 같은 종류의 모든 TV에 사용 가능하다.

어떤 클래스 데이터타입의 객체에 종속되는 기능은 멤버함수(메서드)로 정의하고, 어떤 데이터타입에도 종속되지 않는 기능은 함수로 정의하면 된다.

```python
# 클래스 : 특별한 컬렉션 자료구조! "사용자 정의 컬렉션 타입"

class Student: # 사용자 정의 클래스를 만들때는 "대문자"로 시작

    # 생성자 메서드
    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    # 기본 메서드
    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2

    def print_info(self):
        print(f"이름 : {self.name}")
        print(f"국어 : {self.kor}")
        print(f"영어 : {self.eng}")
        print(f"총점 : {self.total()}")
        print(f"평균 : {self.average()}")
```

```python
s1 = Student("홍길동", 100, 90) # Student 클래스 안의 __init__을 호출함 → Student.__init__(self, name, kor, eng)

print(s1.name)
print(s1.kor)
print(s1.eng)

print("-"*15)
s1.print_info()
print("-"*15)

s2 = Student("김철수", 99, 88)
s2.print_info()
```

```python
class Account:

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, money):
        self.balance += money

    def withdraw(self, money):
        self.balance -= money

    def print_info(self):
        print(f"예금주 : {self.owner}")
        print(f"잔액 : {self.balance}원")
```

```python
acc = Account("홍길동", 10000)

acc.deposit(5000)
acc.withdraw(3000)

acc.print_info()
```

```python
class Car:

    def __init__(self, model):
        self.model = model
        self.speed = 0

    def accelerate(self):
        self.speed += 10

    def brake(self):
        self.speed -= 10

    def print_info(self):
        print(f"차량 : {self.model}")
        print(f"속도 : {self.speed:,}km/h")
```

```python
car = Car("소나타")

car.accelerate()
car.accelerate()
car.brake()

car.print_info()
```

```python
class TV:

    def __init__(self):
        self.power = False
        self.channel = 1

    def on(self):
        self.power = True

    def channel_up(self):
        self.channel += 1

    def print_info(self):
        print(f"전원 : {self.power}")
        print(f"채널 : {self.channel}")

tv = TV()

tv.on()
tv.channel_up()
tv.channel_up()

tv.print_info()
```

```python
class Rectangle:

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self): # 넓이
        return self.width * self.height

    def perimeter(self): # 둘레
        return (self.width + self.height) * 2


r = Rectangle(10, 5)

print(r.area())
print(r.perimeter())
```
