```python
# 리스트에서 리스트 추출 : 컴프리헨션

# 계산에서 사용되는 임시 데이터셋(원본X)
values = [r for r in sales if isinstance(r["판매액"], int)] # 결측치 제거
len(values) # 18
print(values)
```

```python
# 1. 최댓값 / 최솟값 찾기 : 판매액이 가장 큰 거래와 가장 작은 거래를 찾으시오.

max_sale = max(values, key = lambda x : x["판매액"]) # 고차함수, 람다
min_sale = min(values, key = lambda x : x["판매액"])

print(f"최대값 : {max_sale}")
print(f"최소값 : {min_sale}")
```

```python
# 2. Top-N 추출 : 판매액이 가장 큰 거래 Top3를 출력하시오.

sort_values = sorted(values, key = lambda x : x["판매액"], reverse=True)

# print(sort_values)

rank_num = 1
for r in sort_values[:3]:
    print(f"{rank_num}위 : {r}")
    rank_num += 1
```

```python
# 3. 빈도수 계산 : 상품별 판매 건수를 구하시오.

count = {}

for r in sales:
    key_name = r["상품"]

    if key_name not in count:
        count[key_name] = 0
        
    count[key_name] += 1

print(count)

for k, v in count.items():
    print(f"{k} : {v}건")
```

```python
# 4. 그룹별 집계(Group By) : 지역별 총 판매액을 구하시오.

local_sum = {}

for r in sales:
    price = r["판매액"]
    
    if not isinstance(price, int): # 결측치 이면
        continue # 다음 반복으로 건너뛰기
    
    key_name = r["지역"]
    if key_name not in local_sum:
        local_sum[key_name] = 0

    local_sum[key_name] += price

print(local_sum)
for k, v in local_sum.items():
    print(f"{k} : {v:,}원")
```

```python
# 5. 중복 제거 : 판매된 지역 목록만 출력하시오.

regions = set()

for r in sales:
    regions.add(r["지역"])

print(regions)
```

```python
# 8. 이동평균(Moving Average) : 판매액 슬라이딩 윈도우, 3일 이동평균 계산

prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# print(prices)

window = 3

for i in range(0, len(prices) - (window-1)):
    total = sum(prices[i: i + window])
    avg = round(total / window, 3)
    print(f"{sales[i]['날짜']} ~ {sales[i + window - 1]['날짜']} 평균 판매액 : {avg:,}원")
```

```python
# 9. 누적평균(Running Average)

prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# print(prices)
prefix = [0]

for p in prices:
    prefix.append(prefix[-1] + p)

print(prefix)

run_avgs = []

for i in range(1, len(prefix)):
    run_avgs.append(prefix[i] / i)

    avg = run_avgs[-1]

    if i == 1:
        print(f"{i}일차 평균 : {avg:,.3f}원")
    else:
        print(f"1~{i}일차 평균 : {avg:,.3f}원")
```

```python
# 10. 이상치 탐지(Threshold) : 9999999를 찾아라. 또는 200만원 이상을 이상치로 판단.

for r in sales:
    if isinstance(r["판매액"], int):
        if r["판매액"] >= 2000000:
            print("이상치 발견", r)
```

```python
# 11. 결측치 처리 : "", "결측"을 중앙값으로 변경.

avg = sum(prices) / len(prices) # 평균

# 중앙값 정렬
# print(len(prices)) : 짝수
prices.sort()
print(prices)

if len(prices) % 2 == 1: # 홀수 이면
    median = prices[len(prices // 2)] # 중앙값 1개
else: # 짝수 이면
    median = (prices[len(prices) // 2] + prices[len(prices) // 2 - 1]) / 2

print(median)
print("원본", sales)

for r in sales:
    if not isinstance(r["판매액"], int):
        r["판매액"] = median

print("결측치 처리", sales)
```

```python
# 12. 범주별 통계 : 상품별 평균/ 최대/ 최소 판매액, 총 판매량

result = {}

for r in sales:
    p_name = r["상품"]
    price = r["판매액"]
    total_cnt = r["수량"]

    if r["상품"] not in result:
        result[p_name] = {"합계" : 0, "건수" : 0, "최대" : price, "최소" : price, "총수량" : 0}
        # {
        #   "키보드" : {"합계" : 0, "최대" : 0, "최소" : 0, "수량" : 0},
        #   "모니터" : {"합계" : 0, "최대" : 0, "최소" : 0, "수량" : 0},
        #    ...
        # }
    
    info = result[p_name]
    
    info["합계"] += price
    info["건수"] += 1
    info["총수량"] += total_cnt

    # 최대 / 최소
    # info["최대"] = result["노트북"][최대] 
    if info["최대"] < price:
        info["최대"] = price
    
    if info["최소"] > price:
        info["최소"] = price

result
        
```
