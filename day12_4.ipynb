< 큐(Queue) >

- 스택은 뒤에서 넣고 뒤에서 꺼낸다(LIFO)
- 큐는 뒤에서 넣고 앞에서 꺼낸다(FIFO)

1. 큐(Queue)의 개념
FIFO(First In First Out) 먼저 들어온 데이터가 먼저 나온다.

```python
# 리스트로 큐 구현

# 데이터 넣기(Enqueue)

queue = []

queue.append("A")
queue.append("B")
queue.append("C")

print(queue)
```

```python
# 데이터 꺼내기(Dequeue)

queue = ["A", "B", "C"]

print(queue.pop(0))
print(queue)

print(queue.pop(0))
print(queue)

print(queue.pop(0))
print(queue)
```

```python
queue = ["A", "B", "C"]
pop_index = 0

while pop_index < len(queue):
    print(queue[pop_index])
    print(queue[pop_index + 1:])
    pop_index += 1

print(queue)
```

```python
# 은행 번호표 예제

queue = []

queue.append("홍길동")
queue.append("이순신")
queue.append("김유신")

print("대기 :", queue)
print("입장 :", queue.pop(0))
print("현재 대기 :", queue)
```

```python
# 프린트 출력 큐 예제

queue = []

queue.append("문서1")
queue.append("문서2")
queue.append("문서3")

while len(queue) > 0:
    print("출력 :", queue.pop(0))
```

```python
# 음식 주문 예제

queue = []

queue.append("햄버거")
queue.append("피자")
queue.append("라면")

print("주문 목록 :", queue)

print("조리 시작 :", queue.pop(0))
print("조리 시작 :", queue.pop(0))

print("남은 주문 :", queue)
```

```python
# Stack vs Queue

stack = []

stack.append("A")
stack.append("B")
stack.append("C")

while len(stack) > 0:
    print(stack.pop(), end=" ")
    
queue = []

queue.append("A")
queue.append("B")
queue.append("C")

while len(queue) > 0:
    print(queue.pop(0), end=" ")
```
