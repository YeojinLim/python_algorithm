< 공간 복잡도 >

공간복잡도는 프로그램이 실행되는 동안 사용하는 메모리의 양을 의미한다.
같은 문제를 해결하더라도 어떤 알고리즘은 메모리를 적게 사용하고, 어떤 알고리즘은 많이 사용한다. 공간복잡도를 이해하면 메모리 사용량을 고려하여 더 효율적인 알고리즘을 설계하고 선택할 수 있다.

1. 공간복잡도(Space Complexity)

프로그램이 실행되는 동안 추가로 사용하는 메모리의 양을 공간복잡도라고 한다.
입력이 커질수록 사용하는 메모리가 얼마나 증가하는지 나타낸다. 
보통 Big-O 표기법으로 표현한다. 

예를 들어,
- 변수 1개만 사용하는 프로그램 → 메모리를 거의 사용하지 않음 → O(1)
- 입력 크기만큼 리스트를 만드는 프로그램 → 메모리가 입력 크기에 비례 → O(n)

※ 시간복잡도 vs 공간복잡도
- 시간복잡도 : 얼마나 오래 걸리는가?
- 공간복잡도 : 메모리를 얼마나 사용하는가?

```python
# 1) 공간복잡도 O(1)

n = int(input("숫자 입력 : "))
result = n * 2
print(result)

# 추가로 사용하는 변수는 n,result 딱 2개뿐이다.
# 입력이 10이든 100만이든 변수 개수는 변하지 않는다.
# 공간복잡도 : O(1)
```

```python
# 2) 공간복잡도 O(n)

n = int(input("숫자 개수 : "))

numbers = []

for i in range(n):
    numbers.append(i)

print(numbers)
# O(n)
n = 1000000
numbers = list(range(n))

# 입력(n)이 커질수록 리스트도 커진다.
# 공간복잡도 : O(n)
```

```python
# 3) 공간복잡도 O(n²)

n = int(input("크기 : "))

matrix = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

print(matrix)
# O(n²)
n = 1000
matrix = [[0] * n for _ in range(n)]

# n=3 이면 총 데이터는 [[0,0,0],[0,0,0],[0,0,0]]가 된다. 3*3=9 만큼 메모리가 필요하다.
# 공간복잡도 : O(n²)  
```

1. O(1) : 메모리 사용량이 일정
2. O(n) : 입력 크기만큼 메모리 증가
3. O(n^2) : 입력의 제곱만큼 메모리 증가
