< 덱(Deque, Double Ended Queue) >
알고리즘에서 덱(Deque, Double Ended Queue) 은 양쪽 끝에서 삽입과 삭제가 모두 가능한 자료구조다.

​- append(x) : 오른쪽 삽입
- appendleft(x) : 왼쪽 삽입
- pop() : 오른쪽 삭제
- popleft() : 왼쪽 삭제
- maxlen : 최대 크기 지정

```python
# 덱 생성

from collections import deque
dq = deque()
print(dq)
```

```python
# 오른쪽 데이터 추가 append

from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)
```

```python
# 왼쪽 데이터 추가 appendleft

from collections import deque

dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10)

print(dq)
```

```python
# 오른쪽 삭제 pop

from collections import deque

dq = deque([10, 20, 30])
x = dq.pop()

print(x)
print(dq)
```

```python
# 왼쪽 삭제 popleft

from collections import deque

dq = deque([10, 20, 30])
x = dq.popleft()

print(x)
print(dq)
```

```python
# 양방향 삽입 / 삭제

from collections import deque

dq = deque()

dq.append(10)
dq.append(20)
dq.appendleft(5)
dq.appendleft(1)

print(dq)

print(dq.pop())
print(dq.popleft())

print(dq)
```

```python
# 회전 roate

from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)

print(dq) # deque([5, 1, 2, 3, 4])

dq2 = deque([1, 2, 3, 4, 5])
x = dq2.pop()
dq2.appendleft(x)
print(dq2) # deque([5, 1, 2, 3, 4])

dq.rotate(2) # 우측으로 두칸 이동(회전)
print(dq) # deque([3, 4, 5, 1, 2])
```

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(-2) # 왼쪽으로 두 칸 회전

print(dq)
```

```python
# 덱 자료구조 실습

# 최근 방분 기록 n개 저장

from collections import deque

history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

```python
# 카드 게임(원카드, 우노, ..) 시뮬레이션

from collections import deque

cards = deque([1, 2, 3, 4, 5])

while len(cards) > 1:
    print("버림 :", cards.popleft())
    cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
