< 스택(Stack) >

1. 리스트의 특징과 활용
리스트는 맨 뒤 데이터 추가(append()),  맨 뒤 데이터 삭제(pop()) 가 매우 빠르다. 그래서 스택(Stack) 을 구현하기에 적합하다.

2. 스택(Stack)

1) 구조 : LIFO(Last In First Out), 나중에 들어온 데이터가 가장 먼저 나온다.
2) 저장 : push : 데이터 넣기 -> 리스트의 append() 사용
3) 삭제 : pop : 데이터 꺼내기 -> 리스트의 pop() 사용​
4) Top 확인 : stack[-1]
5) 비어있는지 확인 : len(stack) == 0
6) 대표 활용 : 실행취소(Undo), 웹 브라우저 뒤로 가기, 괄호 검사, 함수 호출

```python
# push
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print(stack)
```

```python
# pop

num = stack.pop()
print(num)
print(stack)

num = stack.pop()
print(num)
print(stack)

num = stack.pop()
print(num)
print(stack)
```

```python
# Top확인

stack = [10, 20, 30]

print(stack[-1]) # Top 확인
print(stack)
```

```python
# 스택이 비었는지 확인

stack = []

if len(stack) == 0:
    print("스택이 비어 있습니다.")
else:
    print("데이터가 있습니다.")
```

```python
# 웹 브라우저 "뒤로가기" 구현

stack = []

stack.append("네이버")
stack.append("구글")
stack.append("ChatGPT")

print("현재 방문 기록 :", stack)
print("뒤로가기 :", stack.pop())
print("현재 페이지 :", stack[-1])
print("뒤로가기 :", stack.pop())
print("현재 페이지 :", stack[-1])
```

```python
# "접시(동전) 쌓기" 구현

stack = []

stack.append("접시1")
stack.append("접시2")
stack.append("접시3")

while len(stack) > 0:
    print(stack.pop())
```

```python
# 100엔 스시집 계산

plates = []
plates.append({"color" : "red", "price" : 100})
plates.append({"color" : "blue", "price" : 150})
plates.append({"color" : "blue", "price" : 150})
plates.append({"color" : "black", "price" : 200})
plates.append({"color" : "red", "price" : 100})

print(plates)

total_price = 0

while len(plates) > 0:
    p = plates.pop()
    total_price += p["price"]

print(f"{total_price:,}엔")
print(plates)
```
