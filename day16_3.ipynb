1. 절차 지향적 프로그래밍 
- 코드 내용 : 객체(데이터) + 함수(기능)

2. 객체 지향적 프로그래밍
- 코드 내용: 객체(데이터 + 매서드(기능))
- 메서드는 멤버함수라고 부르기도 한다.

```python
# 1. 절차지향 프로그래밍
# 절차(순서)대로 코드를 작성한다. -> 데이터를 만들고 그 데이터를 사용(참조)하는 함수를 만들어 호출한다. -> 데이터와 함수가 서로 따로 존재 한다.

student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85
}

total = student["kor"] + student["eng"]
avg = total / 2

print(total)
print(avg)

# 데이터 → 함수 실행 → 결과 출력
```

```python
# 2. 객체지향 프로그래밍

class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2


s = Student("홍길동", 90, 85) # 학생 객체안에 자가 자신의 객체(self)를 다루는 메서드 포함 → 데이터와 기능을 하나로 묶는다. 

print(s.total())
print(s.average())
```
