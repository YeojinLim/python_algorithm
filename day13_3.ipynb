< 트리 자료구조 >

1) 트리 자료구조의 목적
- 계층적인 데이터들을 빠르게 탐색/검색/관리가 가능하한 자료구조
- 트리(이진탐색트리)는 삽입과 삭제에 탐색 과정이 필요하지만(저장할 위치를 찾는 과정이 필요해서 리스트보다 비용이 더 드는 경우가 많다), 데이터를 빠르게 탐색하기 위해 사용하는 자료구조이다.

​

* 선형 구조 vs 트리 구조

- 선형 구조: 앞에서부터 순서대로 탐색
- 트리 구조: 부모-자식 구조를 따라 필요한 데이터만 탐색하여 더 빠르게 찾을 수 있음. 물론 트리가 한쪽으로 치우치면 최악의 경우 선형구조와 동일한 탐색시간이 느려질 수 있다.

```python
# 데이터 분석에서 트리 자료구조
# 사람의 정보
person = {
    "age": 35,
    "income": 6000,
    "credit": 750
}

# 변수에 저장
age = person["age"]
income = person["income"]
credit = person["credit"]

# 사람이 직접 만든 규칙
if age >= 20:
    if income >= 5000:
        if credit >= 700:
            print("승인")
        else:
            print("거절")
    else:
        print("거절")
else:
    print("거절")
```

```python
# if문을 Tree로 표현
# 1. if문

color = "Red"
size = "Large"

if color == "Red":
    if size == "Large":
        print("사과 (Apple)")
    else:
        print("체리 (Cherry)")
elif color == "Purple":
    print("포도 (Grape)")
else:
    print("모르는 과일")
```

```python
# 2. Tree
# 이때 트리 자료구조는 데이터를 저장하기 위한 목적보다는 조건 업무프로세스의 순서를 저장

fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}
```

```python
# 트리 탐색 코드
# isinstance(a,b)->bool 함수는 변수 a가 b타입의 객체(instance)이면 True 아니면 False를 반환하는 함수

# 입력값
fruit = {
    "color": "Red",
    "size": "Small" #"Large" "Small"
}

# 트리 생성
fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}

node = fruit_tree # 현재 위치를 트리의 맨 꼭대기(뿌리)로 설정

# 트리를 탐색하는 반복문
while isinstance(node, dict): # node객체가 dict 이라면 True 아니면 False → 질문이 남아있는 동안 계속 반복

    # 현재 위치의 "question"을 찾아 질문은 key, 정답은 value에 저장 
    key, value = node["question"] 
    # 1 : key, value = ("color", "Red")
    # 2 : key, value = ("size", "Large")
    
    if fruit[key] == value:
        node = node["yes"] # 위치를 "yes" 딕셔너리로 이동
    else:
        node = node["no"] # 위치를 "no" 딕셔너리로 이동

# 반복문을 빠져 나오면 마지막 node값을 출력
print(node)
```

기존 방식: 규칙을 if문으로 직접 작성한다.
트리 방식: 규칙은 tree에 저장하고, 트리 탐색 코드가 if문으로 모든 규칙을 처리한다.
