​
"알고리즘이란 어떤 문제를 해결하기 위한 절차나 방법을 순서대로 정리한 것"이다.
프로그램은 그 알고리즘을 프로그래밍 언어(Python, C, Java 등)로 구현한 것이다.

```python
# 문제 1. 팀장 번호 구하기 (산술 연산자)

# 1번부터 100번까지 학생이 있다. 1~10번은 1팀 / 11~20번은 2팀  / 21~30번은 3팀 ... 각 팀은 10명씩이며, 첫 번째 학생이 팀장이다.
# 학생 번호를 입력받아 해당 학생의 팀장 번호를 출력하는 프로그램을 작성하시오.

num = int(input("학생 번호 :"))
# num

lead_num = ((num-1) // 10 * 10) + 1

print(f"학생 번호 : {num}")
print(f"팀장 번호 : {lead_num}")
```

```python
# 문제 2. 홀수와 짝수 판별하기 (if~else)

# 정수를 입력받아 홀수인지 짝수인지 출력하는 프로그램을 작성하시오.
try:
    num = int(input("숫자를 입력하세요"))

    if num % 2 == 0:
        print(f"숫자 : {num}")
        print("짝수")
    else:
        print(f"숫자 : {num}")
        print("홀수")

except ValueError as e:
    print("숫자만 입력하세요")
```

```python
# 문제 3. 양수, 음수, 0 판별하기 (if~elif~else)

# 정수를 입력받아 양수, 음수 또는 0을 출력하는 프로그램을 작성하시오.

try:
    num = int(input("정수를 입력하세요: "))

    if num > 0:
        print("양수")
    elif num < 0:
        print("음수")
    else:
        print("0")

except ValueError as e:
    print("정수만 입력하세요")
```

```python
# 문제 4. 20대 출입 가능 여부

# 두 사람의 나이를 입력받는다.
# 두 사람 모두 20대(20~29세)이면 출입 가능을 출력하고, 그 외에는 출입 불가를 출력하는 프로그램을 작성하시오.

age1 = int(input("첫 번째 사람의 나이: "))
age2 = int(input("두 번째 사람의 나이: "))

if (20 <= age1 <= 29) and (20 <= age2 <= 29):
    print("출입 가능")
else:
    print("출입 불가")
```

```python
# 문제 5. 1~50 사이의 5의 배수 출력 : 1부터 50까지의 수 중 5의 배수만 출력하는 프로그램을 작성하시오.

for i in range(5, 51, 5): # range(시작, 종료, 증감폭)
    print(i)
```

```python
# 문제 6. 1~10 사이 홀수의 합 구하기 : 1부터 10까지의 홀수의 합을 출력하는 프로그램을 작성하시오.

total = 0

for i in range(1, 11): # 1 ~ 10
    if i % 2 != 0:  # 2로 나눈 나머지가 0이 아니면 홀수
        total += i

print(f"1~10 사이 홀수의 합: {total}")
```

```python
# 문제 7. 윤년 구하기 : 2014년부터 2500년까지의 윤년을 모두 출력하는 프로그램을 작성하시오.

# 윤년의 조건
# - 400으로 나누어 떨어지는 해 
# - 또는 4로 나누어 떨어지지만 100으로는 나누어 떨어지지 않는 해

for year in range(2014, 2501):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        print(f"{year}년", end=" ")
```

```python
# 문제 8. 소수 판별하기 : 정수를 입력받아 소수인지 아닌지 판별하는 프로그램을 작성하시오.

# 소수는 1보다 큰 자연수 중 1과 자기 자신만으로 나누어지는 수이다.

try:
    num = int(input("정수를 입력하세요: "))

    if num < 2:
        print(f"{num}은 소수가 아닙니다.")
    else:
        for i in range(2, num):
            if num % i == 0:
                print(f"{num}은 소수가 아닙니다.")  # 나누어떨어지면 소수가 아님
                break
        else:
            print(f"{num}은 소수입니다.")

except ValueError as e:
    print("정수만 입력하세요")
```

```python
# 문제 9. 3~100 사이의 소수 개수 구하기 : 3부터 100까지의 소수의 개수를 출력하는 프로그램을 작성하시오.

prime_num = []
prime_count = 0

for num in range(3,101):
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        prime_num.append(num)
        prime_count += 1

print(f"3~100 사이의 소수 : {prime_num}")
print(f"3~100 사이의 소수 개수 : {prime_count}개")
```

```python
# 문제 9-2. 입력받은 숫자 사이의 소수 개수 구하기 

try:
    prime_num = []
    prime_count = 0

    a = int(input("정수를 입력하세요: "))
    b = int(input("정수를 입력하세요: "))

    for num in range(a, b+1):
        prime = True

        for i in range(2, num):
            if num % i == 0:
                prime = False
                break
        if prime:
            prime_num.append(num)
            prime_count += 1

    print(f"{a}~{b} 사이의 소수 : {prime_num}")
    print(f"{a}~{b} 사이의 소수 개수 : {prime_count}개")

except ValueError as e:
    print("정수만 입력하세요")
```

```python
# 문제 10. 별 출력하기

for i in range(1, 6):
    print("*" * i)
```

```python
# 문제 11. X 모양 별 출력

for i in range(10):
    for j in range(10):
        if i == j or i + j == 9:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
# 12. 가로 구구단 출력 : 2단부터 9단까지를 가로 방향으로 출력하는 프로그램을 작성하시오.

for dan in range(2, 10): # 2단 ~ 9단
    for i in range(1, 10): # 1 ~ 9
        print(f"{dan}*{i}={i*dan:2d}", end=" ") # {:2d} : 숫자를 2자리 공간에 맞춰 출력
    print()
```

```python
# 13. 세로 구구단 출력 : 2단부터 9단까지를 세로 방향으로 출력하는 프로그램을 작성하시오.

for i in range(1, 10):  # 1 ~ 9
    for dan in range(2, 10):  # 2단 ~ 9단
        print(f"{dan}*{i}={dan*i:2d}", end=" ")
    print()  # 한 행이 끝나면 줄바꿈
```
